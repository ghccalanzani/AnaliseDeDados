```python
import pandas as pd
import numpy as np
```

```python
index = ['Firefox', 'Chrome', 'Safari', 'IE10', 'Konqueror']

navegadores = pd.DataFrame({'http_status': [200, 200, 404, 404, 301],
                            'response_time': [0.04, 0.02, 0.07, 0.08, 1.0]},
                            index=index)

navegadores
```

```python
###### Questao 1 #########
#(a) Realize um filtro a partir da coluna http_status onde o código é igual à 200.
navegadores_filtro200 = navegadores[navegadores['http_status'] == 200]
navegadores_filtro200
```

```python
#(b) Realize um filtro a partir da coluna response_time onde o tempo de resposta é menor que 0.08 ms
navegadores_filtroMenor008 = navegadores[navegadores['response_time'] < 0.08]
navegadores_filtroMenor008
```

```python
#(c) Reindexe o DataFrame com o objeto new_index apresentado logo abaixo da criação do objeto navegadores.
new_index = ['Safari', 'Iceweasel', 'Comodo Dragon', 'IE10', 'Chrome']
navegadoresReindex = navegadores.reindex(new_index, fill_value=0).copy()
navegadoresReindex
```

```python
#(d) Refaça as atividades A e B e verifique se os resultados são os mesmos.
navegadores_filtro200Reindex = navegadoresReindex[navegadoresReindex['http_status'] == 200]
navegadores_filtroMenor008Reindex = navegadoresReindex[navegadoresReindex['response_time'] < 0.08]

navegadores_filtro200Reindex
```

```python
navegadores_filtroMenor008Reindex
```

```python
#(e) Substitua o campo http_status do browser Iceweasel pelo código 501 e o campo response_time por 10.0.
navegadoresReindex.loc['Iceweasel', 'http_status'] = 501
navegadoresReindex.loc['Iceweasel', 'response_time'] = 10.0
navegadoresReindex
```

```python
#(f) Substitua o campo http_status do browser Comodo Drago pelo código 503 e o campo response_time por 8.0
navegadoresReindex.loc['Comodo Dragon', 'http_status'] = 503
navegadoresReindex.loc['Comodo Dragon', 'response_time'] = 8.0
navegadoresReindex
```

```python
#(g) Com a nova reindexação, alguns índices acabaram sendo removidos. Descubra quais foram removidos e reinsira-os com seus respectivos campos.
removed_index = navegadores.index.difference(navegadoresReindex.index)

adicionarNavegadores = pd.DataFrame({
                       'http_status': [404, 200],
                       'response_time': [0.07, 0.02]
                       }, index=removed_index)

navegadoresReindex = pd.concat([navegadoresReindex, adicionarNavegadores])
navegadoresReindex
```

```python
#(h) Pesquise as versões atuais de cada navegador e inclua um novo atributo chamado version com as informações encontradas.
navegadoresReindex.loc['Safari', 'version'] = 1
navegadoresReindex.loc['Iceweasel', 'version'] = 2
navegadoresReindex.loc['Comodo Dragon', 'version'] = 3
navegadoresReindex.loc['IE10', 'version'] = 4
navegadoresReindex.loc['Chrome', 'version'] = 5
navegadoresReindex.loc['Firefox', 'version'] = 6
navegadoresReindex.loc['Konqueror', 'version'] = 7
navegadoresReindex
```

```python
###### Questao 2 #########
# Pesquise e crie um exemplo sobre os seguintes métodos do DataFrame Pandas
#(a) dropna e drop_duplicates

dfNovo = pd.DataFrame({
    'brand': ['Yum Yum', 'Yum Yum', 'Indomie', 'Indomie', 'Indomie'],
    'style': ['cup', 'cup', 'cup', 'pack', 'pack'],
    'rating': [4, 4, 3.5, 15, np.nan]
})
dfNovo
```

```python
dfNovo = dfNovo.dropna()
dfNovo = dfNovo.drop_duplicates()
dfNovo
```

```python
#(b) merge e concat

df1 = pd.DataFrame({'lkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [1, 2, 3, 5]})
df2 = pd.DataFrame({'rkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [5, 6, 7, 8]})

df_merge = pd.merge(df1, df2, left_on='lkey', right_on='rkey')
df_merge
```

```python
df_concat = pd.concat([df1, df2])
df_concat
```

```python
#(c) groupby

dfC = pd.DataFrame({'Animal': ['Falcon', 'Falcon',
                              'Parrot', 'Parrot'],
                   'Max Speed': [380., 370., 24., 26.]})
dfC
```
