```python
# Calcule a mÃ©dia das idades dos passageiros
media = titanic['Age'].mean()
print("Media = ", media)
```

```python
# Calcule a mediana da idade e da taxa paga
medianaIdade = titanic['Age'].median()
medianaTaxa = titanic['Fare'].median()
print("Mediana Idade = ", medianaIdade, "\nMediana Taxa = ", medianaTaxa)
```
