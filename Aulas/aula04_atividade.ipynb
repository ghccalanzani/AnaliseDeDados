```python
#19 - Usando o PANDAS, responda:
#a) Qual a média de temperatura por estação do ano?
dfS1 = df.loc[df['season'] == 1]
print(dfS1['temp'].mean())
dfS2 = df.loc[df['season'] == 2]
print(dfS2['temp'].mean())
dfS3 = df.loc[df['season'] == 3]
print(dfS3['temp'].mean())
dfS4 = df.loc[df['season'] == 4]
print(dfS4['temp'].mean())
```

```python
#b) Qual a média de sensação térmica por estação do ano?
dfS1ST = df.loc[df['season'] == 1]
print(dfS1ST['atemp'].mean())
dfS2ST = df.loc[df['season'] == 2]
print(dfS2ST['atemp'].mean())
dfS3ST = df.loc[df['season'] == 3]
print(dfS3ST['atemp'].mean())
dfS4ST = df.loc[df['season'] == 4]
print(dfS4ST['atemp'].mean())
```

```python
#c) Qual a estação do ano que mais possui aluguéis?
dfS1 = df.loc[df['season'] == 1]
print(dfS1['holiday'].sum())
dfS2 = df.loc[df['season'] == 2]
print(dfS2['holiday'].sum())
dfS3 = df.loc[df['season'] == 3]
print(dfS3['holiday'].sum())
dfS4 = df.loc[df['season'] == 4]
print(dfS4['holiday'].sum())
print("A estacao que possui mais alugueis são: 3 e 4")
```

```python
#d) Qual o clima do ano que mais possui aluguéis?
dfC1 = df.loc[df['weather'] == 1]
print(dfC1['count'].sum())
dfC2 = df.loc[df['weather'] == 2]
print(dfC2['count'].sum())
dfC3 = df.loc[df['weather'] == 3]
print(dfC3['count'].sum())
dfC4 = df.loc[df['weather'] == 4]
print(dfC4['count'].sum())
print("O clima do ano que possui mais alugueis é o Clima 1")
```
